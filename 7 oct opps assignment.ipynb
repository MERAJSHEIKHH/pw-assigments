# Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.


```python
# self






class Book:
    def __init__(self, title, available=True):
        self.title = title
        self.available = available

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def borrow_book(self, title):
        for book in self.books:
            if book.title == title and book.available:
                book.available = False
                return f"The book '{title}' has been borrowed."
            else:
                
                return f"The book '{title}' is either unavailable or does not exist in the library."

    def display_available_books(self):
        available_books = [book.title for book in self.books if book.available]
        return available_books

# Create a library object
library = Library()

# Add books to the library
library.add_book(Book("The Catcher in the Rye"))
library.add_book(Book("To Kill a Mockingbird"))
library.add_book(Book("1984"))
library.add_book(Book("Pride and Prejudice"))


# Borrow a book

print(library.borrow_book("To Kill a Mockingbird"))



# Display available books

print(library.display_available_books())




# Try to borrow a book that is not available

print(library.borrow_book("To Kill a Mockingbird"))


# Return the borrowed book

print(library.borrow_book("The Catcher in the Rye"))


# Display available books after return

print(library.display_available_books())



# Display available books

print(library.display_available_books())
```

# Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.


```python
class Product:
    def __init__(self, product_id, name, quantity):
        self.product_id = product_id
        self.name = name
        self.quantity = quantity

class InventorySystem:
    def __init__(self):
        self.inventory = []

    def add_product(self, product):
        self.inventory.append(product)

    def update_quantity(self, product_id, new_quantity):
        for product in self.inventory:
            if product.product_id == product_id:
                product.quantity = new_quantity
                return f"Quantity for {product.name} updated to {new_quantity}."
        return f"Product with ID {product_id} not found in inventory."

    def display_available_products(self):
        available_products = [(product.name, product.quantity) for product in self.inventory if product.quantity > 0]
        return available_products

# Example Usage
inventory_system = InventorySystem()

# Add products to inventory
inventory_system.add_product(Product(1, "Product A", 10))
inventory_system.add_product(Product(2, "Product B", 5))
inventory_system.add_product(Product(3, "Product C", 8))

#Display available products
print(inventory_system.display_available_products())

# Update quantity of a product
print(inventory_system.update_quantity(2, 7))

# Display available products after update
print(inventory_system.display_available_products())
```

# Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.


```python
class shape:
    def __init__(self,length,width,height):
        self.length= length
        self.width=width
        self.height=height
        
class rectangle(shape):
    
    def cal_area(self):
        return self.length*self.width
    
    def cal_perimeter(self):
        return 2*(self.length+self.width)
    

rectangle= rectangle(20,10,5) 

print(f"the area of rectangle is {rectangle.cal_area()}")
print(f"the perimeter of rectasngle is {rectangle.cal_perimeter()}")


```
